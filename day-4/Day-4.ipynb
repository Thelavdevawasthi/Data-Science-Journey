#Set & Set methods

```python
s = {2,5,8,"hari",8}
s.add(45)    #add one
s.update([4,5,9,11])
s.pop()     #remove random element
s
```

```python
#Set Operations
a={1,2,3,4}
b = {4,5,7,8}
a.union(b)
a.intersection(b)
a.difference(b)   #difference in a
a.issubset(b)
```

#File Handling

```python
f = open("text.txt", "r") #open file in read mode
f.read()                  #read file
f.close()                 #close file
```

```python
#write mode
l = "Learnign Data Science Day-4"
f = open("text.txt", "w")
f.write(l)
f.close()
```

```python
#append
a = ("Day-4 File handling")
f = open("text.txt", "a")
f.write(a)
f.close()
```

```python
#with open() ensures the file is automatically closed after execution
with open ("text.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#OS module to Checking if a File Exists
import os 
if os.path.exists("text.txt"):
    print("File already exists")
else:
    print("File doesn't exists")
```

```python
#Deleting a File
import os
if os.path.exists("sample.txt"):
    os.remove("sample.txt")
    print("File deleted")
else:
    print("file doesn't exists")
    
```

#JSON module

```python
import json
data = {"name": "lavdev", "age": 25, "city": "Kathmandu"}
record = json.dumps(data)       #Convert Python object to JSON string
print(record)  
```

```python
#Read JSON data from a file
with open("sample.json", "r") as file:
    data = json.load(file)      #read json data
print(data)
# Format json data

formatted =json.dumps(data, indent=5)
print(formatted)
```
