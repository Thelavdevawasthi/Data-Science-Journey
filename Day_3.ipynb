#Loops

```python
#for loop
l = [2,5,6,5,8]
for values in l:
    print(values)
```

```python
#Range number print
for number in range(11,15):
    print(number)
```

```python
#while loop
even = 2
while even < 10:
    print(even)
    even+=2
```

```python
#break 
li = ["mango", "apple", "potato"]
for item in li:
    if (item == "apple"):
        break    #stop here
    print(item)
#Continue
L = ["Nepal", "India", "China"]
for item in L:
    if (item == "India"):
        continue   #skip this/ continue without this
    print(item)  
```
