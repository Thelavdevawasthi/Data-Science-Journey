```python
with open("insta_data.txt", encoding='utf-8') as f: 
    data = f.read()
    #breaking user data into individual profile
chunks = data.split("\n\n")
chunks = [item for item in chunks if len(item)>3]

#storing individual profile data in standard format
def store_data(chunk):
    
    chunk = chunk.strip()    #remove space from both side
    lines = chunk.split("\n")
    user_name = lines[0].replace("username: ",  "")
    name = lines[1].replace("name: ",  "")
    post = lines[2].replace("post: ", "")
    follower = lines[3].replace("follower: ", "")
    following = lines[4].replace("following: ", "")
    bio = lines[5].replace("bio: ", "")
  #  print(name, post, follower, following, bio, sep= "\n")
#returning dictionary
    return {"username": user_name, "name": name, "post": post, "follower": follower, "following": following, "bio": bio}
    
store_data(chunks[6])
```
