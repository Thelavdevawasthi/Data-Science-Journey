#Conditional Statements

```python
#Conditional Statements
z = 16
if z>18:
    print("Eligible for voting")
else:
    print("Not eligible")
```

```python
x = 5
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but not more than 10")
elif x == 5:
    print("x is exactly 5")
else:
    print("x is less than 5")
```

#Match Case

```python
#Match Case -- Similar to switch in C
def http(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Page not found"
        case 500:
            return "server error"
        case _:
            return "unknown status"
print(http(404))
```

#String Formatting and f-strings

```python
name = "Lavdev"
age = 25
print("My name is {} and I am {} years old.".format(name, age))
```

```python
print(f"My name is {name} and I am {age} years old.")   #f string
```

```python
#Formatting Numbers
p = 3.141
print(f"Pi rounded form {p:.2f}")    #Round of 2 values after floating point
```

```python
#Padding and Alignment
print(f"{'Lavdev':<10}")  # Left-align
print(f"{'Lavdev':>10}")  # Right-align
print(f"{'Lavdev':^10}")  # Center-align
```
