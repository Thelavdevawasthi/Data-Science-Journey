```python
import json

def load_data(filename):
    with open(filename, 'r') as f:
        return json.load(f)

def page_may_know(user_id, data):
    user_page = {}
    for page in data["users"]:
        user_page[page['id']] = set(page['liked_pages'])
        
    if user_id not in user_page:
        return []
            
    liked_pages = user_page[user_id]  
    page_suggestion = {}

    # Fixed: added the 's' to .items()
    for other_user, other_liked_pages in user_page.items():
        if other_user != user_id:
            # Fixed: checking intersection with the other user's pages
            shared_pages = liked_pages.intersection(other_liked_pages)
            
            # Recommend pages if there's an overlap
            for page in other_liked_pages:
                if page not in liked_pages:
                    # Fixed: tracks score by specific page id instead of the string 'get'
                    page_suggestion[page] = page_suggestion.get(page, 0) + len(shared_pages)

    # Fixed: sorted_page matches the variable returned on the next line
    sorted_page = sorted(page_suggestion.items(), key=lambda x: x[1], reverse=True)
    return [page_id for page_id, _ in sorted_page] 
        
data = load_data("page_data.json")
user_id = 1
page_recommendation = page_may_know(user_id, data)
print(page_recommendation)
```
